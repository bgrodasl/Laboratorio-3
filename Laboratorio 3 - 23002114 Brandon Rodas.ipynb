### a. ¿Cuál es el segundo municipio más poblado?
### b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# ¿Cuál es el segundo municipio más poblado?
# Primero realizamos un orden de la población,
poblacion_ordenada = dataset.sort_values("Poblacion", ascending=True)

# Luego, creamos la gráfica de barras,
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
skew = round(dataset['Poblacion'].skew(), 4)
kurt = round(dataset['Poblacion'].kurt(), 4)
dataset['Poblacion'].hist(bins=30,color ='blue', density= True, cumulative=False, histtype='stepfilled')
plt.title(f"Poblacion: sk{skew} - kurt={kurt}")
plt.xlabel("Municipio")
plt.ylabel("Poblacion")
plt.xticks(rotation=90)
plt.show()
```

```python
# a. ¿Cuál es la segunda ciudad más poblada?
segundo_municipio = poblacion_ordenada.iloc[-2, 0]
print(f"Segundo municipio más poblado: {segundo_municipio}")
```

```python
# b. ¿Que posición ocupada Granada en el ranking de las ciudad más pobladas?

# Realizamos la agrupación de los datos por provincia y luego una suma de la población,
poblacion_por_provincia = dataset.groupby("Provincia")["Poblacion"].sum()

# Posición de Granada
posicion_granada = poblacion_por_provincia.sort_values(ascending=False).index.get_loc("Granada") + 1

# Imprimir la posición
print(f"Posición de Granada en el ranking: {posicion_granada}")
```

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
!pip install geopy
```

```python
import geopy
import pandas as pd

# Renombra las columnas si es necesario
dataset.rename(columns={"Y": "Latitud", "X": "Longitud"}, inplace=True)

# Filtra la provincia de Madrid (asumiendo que 'Provincia' es la columna)
provincia_madrid = "Madrid"
dataset = dataset[dataset["Provincia"] != provincia_madrid]

from geopy.distance import distance
from geopy.geocoders import Nominatim

madrid_lat = 40.416775
madrid_lon = -3.70379

user_agent = "nombre_de_tu_aplicacion"  # Reemplaza con el nombre de tu aplicación

def calcular_distancia(latitud, longitud):
    madrid_point = geopy.Point(madrid_lat, madrid_lon)
    municipio_point = geopy.Point(latitud, longitud)

    geolocator = Nominatim(user_agent=user_agent)

    madrid_location = geolocator.reverse(madrid_point)
    municipio_location = geolocator.reverse(municipio_point)

    # Acceder a las coordenadas desde los objetos Location
    madrid_lat = madrid_location.latitude
    madrid_lon = madrid_location.longitude
    municipio_lat = municipio_location.latitude
    municipio_lon = municipio_location.longitude

    # Calcular la distancia usando geopy.distance.distance
    return distance((madrid_lat, madrid_lon), (municipio_lat, municipio_lon))

# Calcula las distancias y agrega una nueva columna
dataset["Distancia_Madrid"] = dataset.apply(lambda row: calcular_distancia(row["Latitud"], row["Longitud"]), axis=1)

# Clasifica las distancias usando qcut
dataset["Clasificacion"] = pd.qcut(dataset["Distancia_Madrid"], 5, labels=["Muy Cercano", "Cercano", "Medio", "Lejano", "Muy Lejano"])
```
